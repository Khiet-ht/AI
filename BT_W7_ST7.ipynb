```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt

url = 'iris.csv'
data = pd.read_csv(url)

data.hist()
plt.show()

#moi tuong quan giua cac bien
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot=True)
sns.pairplot(data)

data.head()
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['sepal.length']),y=data['petal.width'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X=pd.DataFrame(data['sepal.length']),y=data['petal.width'])*100,'%')

y_pred = model.predict(X=pd.DataFrame(data['sepal.length']))
data.plot(kind='scatter',x='sepal.length',y='petal.width',figsize=(9,9),color='red',xlim=(1,10))
plt.plot(data['sepal.length'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['sepal.length'],data['sepal.length']**2,data['sepal.length']**3]).T
model.fit(X,y=data['petal.width'])  
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X,y=data['petal.width'])*100,'%')

range = np.arange(4,8,0.1)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='sepal.length',y='petal.width',figsize=(9,9),color='red',xlim=(3,8))
plt.plot(range,y_pred,color='blue')
```
