```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt

url = 'Google_Stock_Price_Train.csv'
data = pd.read_csv(url)

data.hist()
plt.show()

#moi tuong quan giua cac bien
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot=True)
sns.pairplot(data)

data.head()
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['High']), y = data['Open'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
print('He so chinh xac la: ',model.score(X=pd.DataFrame(data['High']), y = data['Open'])*100,'%')
```

```python
y_pred = model.predict(X=pd.DataFrame(data['High']))
data.plot(kind='scatter',x='High',y='Open',figsize=(9,9),color='red',xlim=(250,850))
plt.plot(data['High'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['High'],data['High']**2,data['High']**3]).T
model.fit(X,y=data['Open'])  
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X,y=data['Open'])*100,'%')

range = np.arange(100,900.1,1)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='High',y='Open',figsize=(9,9),color='red',xlim=(250,850))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['High'],data['High']**2,data['High']**3,data['High']**4,data['High']**5]).T
model.fit(X,y=data['Open'])  
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X,y=data['Open'])*100,'%')

range = np.arange(100,900.1,1)
X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='High',y='Open',figsize=(9,9),color='red',xlim=(250,850))
plt.plot(range,y_pred,color='blue')
```
