```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt

url = 'crabs.csv'
data = pd.read_csv(url)

data.hist()
plt.show()

#moi tuong quan giua cac bien
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot=True)
sns.pairplot(data)

data.head()
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['length']),y=data['depth'])
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X=pd.DataFrame(data['length']),y=data['depth'])*100,'%')

```

```python
y_pred = model.predict(X=pd.DataFrame(data['length']))
data.plot(kind='scatter',x='length',y='depth',figsize=(9,9),color='red',xlim=(0,50))
plt.plot(data['length'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['length'],data['length']**2,data['length']**3]).T
model.fit(X,y=data['depth'])  
print('He so intercept la: ',model.intercept_)
print('He so coeffient la: ',model.coef_)
print("do chinh xac la: ",model.score(X,y=data['depth'])*100,'%')
```

```python
range = np.arange(10,50.1,0.1)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='length',y='depth',figsize=(9,9),color='red',xlim=(0,50))
plt.plot(range,y_pred,color='blue')
```
