```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model

url = '50_Startups.csv'
data = pd.read_csv(url)
data.head(2)

data.hist()
plt.show()

#mi tng quan gia cc bin 'correlation'
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot=True)
sns.pairplot(data)
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
print('He so intercept la :',model.intercept_)
print('He so coeffient la :',model.coef_)

#mpg = intercept x + coeffient

y_pred = model.predict(X=pd.DataFrame(data['R&D Spend']))
     

data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='green', xlim=(0,200000))
plt.plot(data['R&D Spend'],y_pred,color='red')
     
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3]).T
y= data['Profit']
model.fit(X,y)
print('Intercept la: ',model.intercept_)
print('Coeffient la: ',model.coef_)
print('Do chuan xac la:',model.score(X,y)*100,'%')

range = np.arange(0,200001,10)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='red',xlim=(0,200000))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3,data['R&D Spend']**4,data['R&D Spend']**5,
                  data['R&D Spend']**6,data['R&D Spend']**7,data['R&D Spend']**8,data['R&D Spend']**9,data['R&D Spend']**10]).T
model.fit(X, y=data['Profit'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',(model.score(X, y=data['Profit'])*100),'%')
range = np.arange(0,200001,10)
X_range = pd.DataFrame([range, range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=X_range)
data.plot(kind ='scatter',x='R&D Spend',y='Profit',figsize = (9,9),color='Red',xlim=(0,200000))
plt.plot(range,y_pred,color='green')
```
